```python
sns.distplot(df['Growth Rate'],kde=True,bins=40)
plt.xlabel('Growth Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Growth Rate')
plt.show()
```

```python
df['Growth Rate'].fillna(df['Growth Rate'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(11,5))
plt.hist(df["Continent"],bins=10,color="maroon",edgecolor="black")
plt.xlabel("Continent")
plt.ylabel("Frequency")
plt.title("Continent")
plt.grid(True)
plt.show()
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(10,8))

Rank = [36, 138, 34, 213, 203]
Country_Territory = ['Afghanistan', 'Albania', 'Algeria', 'American Samoa', 'Andorra']
World_Population_Percentage = [0.52, 0.04, 0.56, 0.00, 0.00]

# Plot World Population Percentage
plt.plot(Rank, World_Population_Percentage, marker='o', linestyle='-')
plt.xlabel("Rank")
plt.ylabel("World Population Percentage")

# Plot Country/Territory
plt.xticks(Rank, Country_Territory, rotation='vertical')

plt.legend(["World Population Percentage"])
plt.show()
```
